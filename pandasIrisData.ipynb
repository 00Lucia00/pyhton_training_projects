```python
# a project for training the basic pandas EDA commands

#Import av Pandas och numpy
import pandas as pd
import numpy as np

#Läser in data från CSV fil
data = pd.read_csv("iris_data_train.csv")

print(data.info())
print(data.shape)
```

```python
#Skapa en Pandas Serial
serData = pd.Series([1, 2, 3.14, True, np.NaN, None])
print(serData)

#Addera +1 till alla element i Serial
serData +=1
print(serData)
```

```python
#Skapa en Dictionary
MyDictionary = {"First":10, "Second":"Hello World", "Last":True}

#Skapa en serial basserat på Dictionary
dicData = pd.Series(MyDictionary)
```

```python
#Skriv ut keys
print(dicData.keys())

#Skriv ut enskillt värde
print(dicData["Second"])

#Skriv ut en Range
print(serData[1:4])
print(dicData['First':'Last'])

#Skriva ut en lista av nycklar
myKeys = [0, 1, 3, 5]
print(serData[myKeys])
```

```python
#Indexpostinioner i en DataFrame
print(data["sepal_length"].head(20))

#2 sätt att få ut samma sak
print(data.sepal_width.head(20))

#Print ut specifika rader
print(data[10:31])

#Hämta ut mellan specifika kolumner
print(data.loc[:,'sepal_length':'petal_width'])
```

```python
#Skapa en scatter plot av dataSet
ax = data.plot.scatter(x='sepal_length', y='sepal_width', color="Red", label = "Sepal", title="Flowers")
data.plot.scatter(x='petal_length', y='petal_width', color="Blue", ax=ax, label ="petal",xlabel="width", ylabel="height")
```

```python
#Color Dictionary
color = {'Iris-virginica': 'Red', 'Iris-setosa': 'Magenta', 'Iris-versicolor':'purple'}

#Skriv ut Scatter plot med samma färger på positionerna
data.plot.scatter(x='sepal_length',y='sepal_width',c="red")

#Skriv ut Scatter plot med olika färger på positionerna
data.plot.scatter(x='sepal_length',y='sepal_width',c=data['class'].map(color))
```

```python
#Color Dictionary
color = {'Iris-virginica': 'Red', 'Iris-setosa': 'Magenta', 'Iris-versicolor':'purple'}

#Kombinera Sepal och Petal med olika färger
ax = data.plot.scatter(x='sepal_length', y='sepal_width', color=data['class'].map(color), title="Flowers")
data.plot.scatter(x='petal_length', y='petal_width', color=data['class'].map(color), ax=ax, xlabel="width", ylabel="height")
```

```python
#Läser in data från CSV fil
data2 = pd.read_csv("iris_data_train.csv")

#Ta bort onödiga kolumner
data2 = data2.drop(columns="Row")
data2 = data2.drop(columns="class")

data2.info()

#Import från SKLearn
from sklearn.preprocessing import StandardScaler

dataStandard = StandardScaler().fit_transform(data2)

pdDataStandard = pd.DataFrame(dataStandard)

pdDataStandard.plot.scatter(x="sepal_length", y="sepal_width")
```
