# while loop övning

```python
# Skirva in en vikt, samt ange om det är i Kg eller Lbs.
# Därefter konvertera till den andra enheten.
vikt = float(input("Skriv in din vikt: "))

while True:
    enhet = input("(K)g eller (L)bs: ").lower()

    if enhet == "k":
        #måste konvertera int till str i en print statement då den inte skriver ut annat än str värden.
        print("Din vikt i Lbs är " + str(vikt / 0.4))
        break # hoppar ur loopen
    elif enhet == "l":
        print("Din vikt i Kg är " + str(vikt * 0.4))
        break
    else:
        print("Felaktig inmatning")
print("Done!") #bara för att indikera att man hoppat ur
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Sets_data/iris_data_train.csv')
```

```python
df.info()
```

```python
ax1 = df.plot.scatter(x='sepal_width',
                      y='sepal_length',
                      c='Red')
```

```python
ax2 = df.plot.scatter(x='petal_width',
                      y='petal_length',
                      c='Blue')
```

```python
ax = df.plot.scatter(x='sepal_width',
                     y='sepal_length',
                     c='Red',
                     label= 'sepal')
df.plot.scatter(x='petal_width',
                y='petal_length',
                c='Blue', 
                label = 'patal', xlabel = 'length', ylabel='width', title = 'Flowers',
                ax=ax)
```
